The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
df = pd.read_csv('data/nobel.csv')

df.head()
```

```python
# most commonly awarded gender and birth country?

top_gender = df['sex'].value_counts().index[0]

top_gender

top_country = df['birth_country'].value_counts().index[0]
top_country
```

```python
# Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

df['decade'] = (df['year'] // 10) * 10
```

```python
filt = df['birth_country']=='United States of America'

df_usa = df[filt]

df_usa.head()
```

```python
total_decade = df.groupby('decade')['prize'].count().reset_index()

total_usa_decade = df_usa.groupby('decade')['prize'].count().reset_index()

decade_agg = total_decade.merge(total_usa_decade,on='decade',how='inner', suffixes=('_total','_usa'))

decade_agg['ratio'] = decade_agg['prize_usa'] / decade_agg['prize_total']

filt = decade_agg['ratio']==decade_agg['ratio'].max()

max_decade_usa = decade_agg.loc[decade_agg[filt].index[0],'decade']

max_decade_usa
```

```python
# Which decade and Nobel Prize category combination had the highest proportion of female laureates?

nobel_female = df[df['sex']=='Female'].groupby(['decade','category'])['prize'].count().reset_index()
# nobel_male = df[df['sex']=='Male'].groupby(['decade','category'])['prize'].count().reset_index()
nobel_total = df.groupby(['decade','category'])['prize'].count().reset_index()
```

```python
nobel_prop = nobel_female.merge(nobel_total, on=['decade','category'], suffixes=('_female','_total'))

nobel_prop['prop'] = nobel_prop['prize_female'] / nobel_prop['prize_total']
max_prop = nobel_prop[nobel_prop['prop']==nobel_prop['prop'].max()]['prop'].index.max()
nobel_prop
x = nobel_prop.loc[max_prop,['decade','category']]
max_female_dict = {x[0]:x[1]}
max_female_dict
```

```python
# Who was the first woman to receive a Nobel Prize, and in what category?

filt = df['sex']=='Female'
df_female = df[filt]
female_index = df_female[df_female['year']==df_female['year'].min()].index[0]

first_woman_name = df_female.loc[female_index,'full_name']
first_woman_category = df_female.loc[female_index,'category']

first_woman_name, first_woman_category
```
